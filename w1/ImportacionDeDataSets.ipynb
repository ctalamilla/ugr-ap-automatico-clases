#Diferentes formas de importar archivos

##Activar Google Drive de forma local
Una forma de tener disponibles los archivos correspondientes a los datasets que trabajaremos es subiendolos a nuestro Google Drive.
Una vez subidos, montamos nuestro Google Drive a nuestro entorno de ejecución de Google Colab.

Para montar Google Drive en tu entorno de ejecución Al ejecutarlo te pedirá la autenticación con tu cuenta y otorgar permisos mediante tu cuenta Google


```python
from google.colab import drive
drive.mount('/content/drive')
```
